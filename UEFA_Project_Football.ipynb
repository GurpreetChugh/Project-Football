```python
import pandas as pd
```

```python
# Creating a list of UEFA seasons
seasons = ['2017-2018', '2018-2019','2019-2020','2020-2021','2021-2022']
uefa_summary_seasons =[]

# Generating a list of UEFA summary standing dataframes for all the 5 seasons
for season in seasons:
    base_url = 'https://fbref.com/en/comps/8/'
    final_url = f'{base_url}{season}/{season}-Champions-League-Stats'
    
    try:
        uefa_tables = pd.read_html(final_url)
        uefa_summary_table = pd.DataFrame(uefa_tables[8])
        # Dropping rows with all Nan values and Notes column
        uefa_summary_table = uefa_summary_table.dropna(how='all').drop('Notes', axis = 1).reset_index(drop=True)
        # Extracting the club name from Squad column
        uefa_summary_table['Squad'] = uefa_summary_table['Squad'].map(lambda x: x.split(' ', maxsplit=1)[1])
        uefa_summary_table['Season'] = f'{season}'
        uefa_summary_seasons.append(uefa_summary_table)
        
    except Exception as e:
        print(e)
        
# List uefa_summary_seasons contains the list of summary dataframes of all the 5 seasons
        
# Concating all the seasons summary dataframes in a single dataframe
uefa_all_seasons_sum_df = pd.concat(uefa_summary_seasons, axis=0).reset_index(drop=True)
uefa_all_seasons_sum_df.head()
   
```

```python
#Renaming Rk as Rank and reordering columns
uefa_all_seasons_sum_df = uefa_all_seasons_sum_df.rename(columns={'Rk':'Rank'})
uefa_all_seasons_sum_df['League'] = 'UEFA'
uefa_all_seasons_sum_df = uefa_all_seasons_sum_df[['Season','Rank', 'Squad', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 'xG',
                                                   'xGA', 'xGD', 'xGD/90', 'Attendance', 'Top Team Scorer', 'Goalkeeper']]
uefa_all_seasons_sum_df.head()
```

```python
# url = 'https://fbref.com/en/comps/8/2021-2022/schedule/2021-2022-Champions-League-Scores-and-Fixtures'
uefa_matches_seasons = []

for season in seasons:
    base_url = 'https://fbref.com/en/comps/8/'
    final_url = f'{base_url}{season}/schedule/{season}-Champions-League-Scores-and-Fixtures'
    try:
        uefa_tables = pd.read_html(final_url)
        uefa_matches_table = pd.DataFrame(uefa_tables[0])
        # Dropping rows with all Nan values and Notes column        
        uefa_matches_table = uefa_matches_table.dropna(how='all').drop('Notes', axis = 1).reset_index(drop=True)
        # Extracting the club name from Home and Away columns
        uefa_matches_table['Home'] = uefa_matches_table['Home'].map(lambda x: " ".join(x.split(" ")[:-1]))
        uefa_matches_table['Away'] = uefa_matches_table['Away'].map(lambda x: " ".join(x.split(" ")[1:]))
        uefa_matches_table['Season'] = f'{season}'
        uefa_matches_seasons.append(uefa_matches_table)
        
    except Exception as e:
        print(e)
        
# List uefa_matches_seasons contains the list of matches dataframes of all the 5 seasons
        
# Concating all the seasons matches dataframes in a single dataframe
uefa_all_seasons_matches_df = pd.concat(uefa_matches_seasons, axis=0).reset_index(drop=True)

# Storing each seasons matches dataframe into the respective variables 
# uefa_matches_2018 = uefa_matches_seasons[0]
# uefa_matches_2019 = uefa_matches_seasons[1]
# uefa_matches_2020 = uefa_matches_seasons[2]
# uefa_matches_2021 = uefa_matches_seasons[3]
# uefa_matches_2022 = uefa_matches_seasons[4]
        
```

```python
# While concatenating all seasons, column headers were repeated as rows
rows_to_drop = uefa_all_seasons_matches_df[uefa_all_seasons_matches_df['Score']=='Score'].index
rows_to_drop
```

```python
uefa_all_seasons_matches_df = uefa_all_seasons_matches_df.drop(rows_to_drop, axis = 0)
uefa_all_seasons_matches_df.info()
```

```python
# Splitting Score column into 2 columns for scores for Home Team and Away Team
uefa_all_seasons_matches_df[['Home Score','Away Score']] = uefa_all_seasons_matches_df['Score'].str.split('–', expand=True)
# Renaming  xG and XG.1 columns
uefa_all_seasons_matches_df = uefa_all_seasons_matches_df.rename(columns={'xG': 'xG Home', 'xG.1': 'xG Away'}) 
uefa_all_seasons_matches_df.info()
```

```python
# Converting numerical columns as floats
uefa_all_seasons_matches_df[['xG Home', 'xG Away','Attendance','Home Score','Away Score']]= uefa_all_seasons_matches_df[['xG Home', 'xG Away','Attendance','Home Score','Away Score']].astype(float)
uefa_all_seasons_matches_df.info()
```

```python
#CAlculating expected and actual Goal Difference
uefa_all_seasons_matches_df['xG Diff'] = uefa_all_seasons_matches_df['xG Home']- uefa_all_seasons_matches_df['xG Away']
uefa_all_seasons_matches_df['Actual Diff'] = uefa_all_seasons_matches_df['Home Score']- uefa_all_seasons_matches_df['Away Score']
uefa_all_seasons_matches_df.head()
```

```python
uefa_all_seasons_matches_df['League'] = 'UEFA'
```

```python
# Reordering the columns
uefa_all_seasons_matches_df = uefa_all_seasons_matches_df[['Season','Round', 'Wk', 'Date', 'Time', 'Home', 'Away','Home Score', 
                                                           'Away Score', 'xG Home', 'xG Away', 'xG Diff', 'Actual Diff', 
                                                           'Attendance', 'Venue', 'Referee']]
uefa_all_seasons_matches_df.head()
```
