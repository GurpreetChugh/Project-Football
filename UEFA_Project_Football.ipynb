```python
import pandas as pd
```

```python
# Creating a list of UEFA seasons
uefa_seasons = ['2017-2018', '2018-2019','2019-2020','2020-2021','2021-2022']
uefa_summary_seasons =[]

# Generating a list of UEFA summary standing dataframes for all the 5 seasons
for season in uefa_seasons:
    base_url = 'https://fbref.com/en/comps/8/'
    final_url = f'{base_url}{season}/{season}-Champions-League-Stats'
    
    try:
        uefa_tables = pd.read_html(final_url)
        uefa_summary_table = pd.DataFrame(uefa_tables[8])
        # Dropping rows with all Nan values and Notes column
        uefa_summary_table = uefa_summary_table.dropna(how='all').drop('Notes', axis = 1).reset_index(drop=True)
        # Extracting the club name from Squad column
        uefa_summary_table['Squad'] = uefa_summary_table['Squad'].map(lambda x: x.split(' ', maxsplit=1)[1])
        uefa_summary_table['Season'] = f'{season}'
        uefa_summary_seasons.append(uefa_summary_table)
        
    except Exception as e:
        print(e)
        
# List uefa_summary_seasons contains the list of summary dataframes of all the 5 seasons
        
# Concating all the seasons summary dataframes in a single dataframe
uefa_summary_consolidated = pd.concat(uefa_summary_seasons, axis=0)
uefa_summary_consolidated.head()
   
```

```python
url = 'https://fbref.com/en/comps/8/2021-2022/schedule/2021-2022-Champions-League-Scores-and-Fixtures'
uefa_matches_seasons = []

for season in uefa_seasons:
    base_url = 'https://fbref.com/en/comps/8/'
    final_url = f'{base_url}{season}/schedule/{season}-Champions-League-Scores-and-Fixtures'
    try:
        uefa_tables = pd.read_html(final_url)
        uefa_matches_table = pd.DataFrame(uefa_tables[0])
        # Dropping rows with all Nan values and Notes column        
        uefa_matches_table = uefa_matches_table.dropna(how='all').drop('Notes', axis = 1).reset_index(drop=True)
        # Extracting the club name from Home and Away columns
        uefa_matches_table['Home'] = uefa_matches_table['Home'].map(lambda x: " ".join(x.split(" ")[:-1]))
        uefa_matches_table['Away'] = uefa_matches_table['Away'].map(lambda x: " ".join(x.split(" ")[1:]))
        uefa_matches_table['Season'] = f'{season}'
        uefa_matches_seasons.append(uefa_matches_table)
        
    except Exception as e:
        print(e)
        
# List uefa_matches_seasons contains the list of matches dataframes of all the 5 seasons
        
# Concating all the seasons matches dataframes in a single dataframe
uefa_matches_consolidated = pd.concat(uefa_matches_seasons, axis=0)
uefa_matches_consolidated.head()
        
# Storing each seasons matches dataframe into the respective variables 
# uefa_matches_2018 = uefa_matches_seasons[0]
# uefa_matches_2019 = uefa_matches_seasons[1]
# uefa_matches_2020 = uefa_matches_seasons[2]
# uefa_matches_2021 = uefa_matches_seasons[3]
# uefa_matches_2022 = uefa_matches_seasons[4]
        
```
