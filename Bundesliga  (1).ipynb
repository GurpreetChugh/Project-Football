# Bundesliga Match Data

```python
table_3 = pd.read_html('https://fbref.com/en/comps/20/2017-2018/schedule/2017-2018-Bundesliga-Scores-and-Fixtures')
Bundes_Matches1718 = pd.DataFrame(table_3[0])
```

```python
table_3 = pd.read_html('https://fbref.com/en/comps/20/2018-2019/schedule/2018-2019-Bundesliga-Scores-and-Fixtures')
Bundes_Matches1819 = pd.DataFrame(table_3[0])
```

```python
table_3 = pd.read_html('https://fbref.com/en/comps/20/2019-2020/schedule/2019-2020-Bundesliga-Scores-and-Fixtures')
Bundes_Matches1920 = pd.DataFrame(table_3[0])
```

```python
table_3 = pd.read_html('https://fbref.com/en/comps/20/2020-2021/schedule/2020-2021-Bundesliga-Scores-and-Fixtures')
Bundes_Matches2021 = pd.DataFrame(table_3[0])
```

```python
table_3 = pd.read_html('https://fbref.com/en/comps/20/2021-2022/schedule/2021-2022-Bundesliga-Scores-and-Fixtures')
Bundes_Matches2122 = pd.DataFrame(table_3[0])
```

```python
Bundes_Matches1718 = Bundes_Matches1718.dropna(how='all')
Bundes_Matches1819 = Bundes_Matches1819.dropna(how='all')
Bundes_Matches1920 = Bundes_Matches1920.dropna(how='all')
Bundes_Matches2021 = Bundes_Matches2021.dropna(how='all')
Bundes_Matches2122 = Bundes_Matches2122.dropna(how='all')
```

```python
Bundes_Matches1718 = Bundes_Matches1718.drop(['Match Report', 'Notes'], axis=1)
Bundes_Matches1819 = Bundes_Matches1819.drop(['Match Report', 'Notes'], axis=1)
Bundes_Matches1920 = Bundes_Matches1920.drop(['Match Report', 'Notes'], axis=1)
Bundes_Matches2021 = Bundes_Matches2021.drop(['Match Report', 'Notes'], axis=1)
Bundes_Matches2122 = Bundes_Matches2122.drop(['Match Report', 'Notes'], axis=1)
```

```python
Bundes_Matches1718["Season"] = "2017-18"
Bundes_Matches1819["Season"] = "2018-19"
Bundes_Matches1920["Season"] = "2019-20"
Bundes_Matches2021["Season"] = "2020-21"
Bundes_Matches2122["Season"] = "2021-22"
```

```python
combined_bundesliga_matches = pd.concat([Bundes_Matches1718,
                                        Bundes_Matches1819,
                                       Bundes_Matches1920,
                                       Bundes_Matches2021,
                                       Bundes_Matches2122 ], axis=0)
```

```python
combined_bundesliga_matches[['Home Score','Away Score']] = combined_bundesliga_matches['Score'].str.split('–', expand=True)
```

```python
combined_bundesliga_matches = combined_bundesliga_matches[['Season', 'Wk', 'Date','Time', 'Home', 'Away',
                                                               'Home Score', 'Away Score', 'xG', 'xG.1', 
                                                               'Attendance', 'Venue', 'Referee']]
```

```python
combined_bundesliga_matches  = combined_bundesliga_matches.rename(columns={ 'xG': 'xG Home', 'xG.1': 'xG Away'}) 
```

```python

combined_bundesliga_matches ["xg Diff"] = combined_bundesliga_matches ['xG Home'].astype(float) - combined_bundesliga_matches ['xG Away'].astype(float)
combined_bundesliga_matches ["Actual Diff"] = combined_bundesliga_matches ['Home Score'].astype(float) - combined_bundesliga_matches ['Away Score'].astype(float)
combined_bundesliga_matches['League'] = 'Bundesliga'
combined_bundesliga_matches.head()
```

# Bundesliga Team Wages

```python
Bundes_Merged_1718 = pd.merge(Bundes_Wages1718, Bundes_Table_1718, how='left', on=['Squad'])
Bundes_Merged_1819 = pd.merge(Bundes_Wages1819 , Bundes_Table_1819, how='left', on=['Squad'])
Bundes_Merged_1920 = pd.merge(Bundes_Wages1920, Bundes_Table_1920, how='left', on=['Squad'])
Bundes_Merged_2021 = pd.merge(Bundes_Wages2021, Bundes_Table_2021, how='left', on=['Squad'])
Bundes_Merged_2122 = pd.merge(Bundes_Wages2122, Bundes_Table_2122, how='left', on=['Squad'])

Bundes_Merged_2122.head()
```

```python
Bundes_1718_df = pd.merge(Bundes_Merged_1718, Bundes_TeamAge_1718, how='left', on=['Squad'])
Bundes_1819_df = pd.merge(Bundes_Merged_1819, Bundes_TeamAge_1819, how='left', on=['Squad'])
Bundes_1920_df = pd.merge(Bundes_Merged_1920, Bundes_TeamAge_1920, how='left', on=['Squad'])
Bundes_2021_df = pd.merge(Bundes_Merged_2021, Bundes_TeamAge_2021, how='left', on=['Squad'])
Bundes_2122_df = pd.merge(Bundes_Merged_2122, Bundes_TeamAge_2122, how='left', on=['Squad'])

Bundes_2122_df.head()
```

```python
Bundes_2122_df.columns
```

```python
Bundes_sum_1718 = Bundes_1718_df [['Rk_y', 'Squad', '# Pl_x', 'Age', 'Poss', 'Annual Wages', 'MP_x', 'W', 'D', 'L', 'GF',
                                  'GA', 'Pts', 'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

Bundes_sum_1819 = Bundes_1819_df [['Rk_y', 'Squad', '# Pl_x', 'Age', 'Poss', 'Annual Wages', 'MP_x', 'W', 'D', 'L', 'GF',
                                  'GA', 'Pts', 'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

Bundes_sum_1920 = Bundes_1920_df[['Rk_y', 'Squad', '# Pl_x', 'Age', 'Poss', 'Annual Wages', 'MP_x', 'W', 'D', 'L', 'GF',
                                  'GA', 'Pts', 'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

Bundes_sum_2021 = Bundes_2021_df[['Rk_y', 'Squad', '# Pl_x', 'Age', 'Poss', 'Annual Wages', 'MP_x', 'W', 'D', 'L', 'GF',
                                  'GA', 'Pts', 'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

Bundes_sum_2122 = Bundes_2122_df [['Rk_y', 'Squad', '# Pl_x', 'Age', 'Poss', 'Annual Wages', 'MP_x', 'W', 'D', 'L', 'GF',
                                  'GA', 'Pts', 'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

Bundes_sum_2122.head()
```

```python
 Bundes_sum_1718 = Bundes_sum_1718.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
Bundes_sum_1819 = Bundes_sum_1819.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
Bundes_sum_1920 = Bundes_sum_1920.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
 Bundes_sum_2021 = Bundes_sum_2021.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})

Bundes_sum_2122 = Bundes_sum_2122.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
Bundes_sum_2122.head()
```

```python
Bundes_sum_1718[['pounds', 'dollars']] = Bundes_sum_1718['Annual Wages'].str.split('$', expand=True)
Bundes_sum_1819[['pounds', 'dollars']] = Bundes_sum_1819['Annual Wages'].str.split('$', expand=True)
Bundes_sum_1920[['pounds', 'dollars']] = Bundes_sum_1920['Annual Wages'].str.split('$', expand=True)
Bundes_sum_2021[['pounds', 'dollars']] = Bundes_sum_2021['Annual Wages'].str.split('$', expand=True)
Bundes_sum_2122[['pounds', 'dollars']] = Bundes_sum_2122['Annual Wages'].str.split('$', expand=True)
```

```python
Bundes_sum_1718[['Annual Wages ($)', 'discard']] = Bundes_sum_1718['dollars'].str.split(')', expand=True)
Bundes_sum_1819[['Annual Wages ($)', 'discard']] = Bundes_sum_1819['dollars'].str.split(')', expand=True)
Bundes_sum_1920[['Annual Wages ($)', 'discard']] = Bundes_sum_1920['dollars'].str.split(')', expand=True)
Bundes_sum_2021[['Annual Wages ($)', 'discard']] = Bundes_sum_2021['dollars'].str.split(')', expand=True)
Bundes_sum_2122[['Annual Wages ($)', 'discard']] = Bundes_sum_2122['dollars'].str.split(')', expand=True)
```

```python
# Drop unwanted columns 
Bundes_sum_1718 = Bundes_sum_1718.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
Bundes_sum_1819 = Bundes_sum_1819.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
Bundes_sum_1920 = Bundes_sum_1920.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
Bundes_sum_2021 = Bundes_sum_2021.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
Bundes_sum_2122 = Bundes_sum_2122.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
```

```python
Bundes_sum_1718 = Bundes_sum_1718.sort_values('Pts', ascending=False)
Bundes_sum_1819 = Bundes_sum_1819.sort_values('Pts', ascending=False)
Bundes_sum_1920 = Bundes_sum_1920.sort_values('Pts', ascending=False)
Bundes_sum_2021 = Bundes_sum_2021.sort_values('Pts', ascending=False)
Bundes_sum_2122 = Bundes_sum_2122.sort_values('Pts', ascending=False)
```

```python
Bundes_sum_1718['Season'] = '2017-2018'
Bundes_sum_1819['Season'] = '2018-2019'
Bundes_sum_1920['Season'] = '2019-2020'
Bundes_sum_2021['Season'] = '2020-2021'
Bundes_sum_2122['Season'] = '2021-2022'
```

```python
Bundes_seasonsdata_df = pd.concat([Bundes_sum_1718, Bundes_sum_1819, 
                                   Bundes_sum_1920, Bundes_sum_2021, Bundes_sum_2122 ])
```

```python
Bundes_seasonsdata_df = Bundes_seasonsdata_df[['Season', 'League Rank', 'Squad', '# of Players',
                                              'Avg Age', 'Avg Poss', 'MP', 'W', 'D', 'GF', 'GA',
                                               'Pts', 'Pts/MP', 'xG', 'xGA', 'xGD', 'xGD/90',
                                              'Attendance', 'Annual Wages ($)']]
Bundes_seasonsdata_df = Bundes_seasonsdata_df.reset_index()
Bundes_seasonsdata_df.head()
```
