```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Get summary tables for each of the last 5 seasons
table = pd.read_html('https://fbref.com/en/comps/9/2017-2018/2017-2018-Premier-League-Stats')
EPL_2018_summary_df = pd.DataFrame(table[0])
```

```python
# Get 2nd summary table (index[2])for each season
table_1 = pd.read_html('https://fbref.com/en/comps/9/2017-2018/2017-2018-Premier-League-Stats')
EPL_2018_summary2_df = pd.DataFrame(table_1[2])
EPL_2018_summary2_df.head()
```

```python
# Remove upper level to allow for merging with other tables
EPL_2018_summary2_df = EPL_2018_summary2_df.droplevel(0, axis = 1)
```

```python
# Get wages table for each of the last 5 seasons
EPL_wages_2018_table = pd.read_html('https://fbref.com/en/comps/9/2017-2018/wages/2017-2018-Premier-League-Wages')
EPL_2018_wages_df = pd.DataFrame(EPL_wages_2018_table[0])
```

```python
# Merge wages and season summary for each season
EPL_2018_merged_df = pd.merge(EPL_2018_wages_df, EPL_2018_summary_df, how="left", on=['Squad'])
EPL_2018_merged_df.head()
```

```python
# Merge in summary table#2 for each season
EPL_2018_complete_df = pd.merge(EPL_2018_merged_df, EPL_2018_summary2_df, how="left", on=['Squad'])
EPL_2018_complete_df.head()
```

```python
#Get columns
EPL_2018_complete_df.columns
```

```python
# Select desired columns for final summary table
EPL_2018_sum_complete_df = EPL_2018_complete_df[['Rk_y', 'Squad','# Pl_x', 'Age', 'Poss', 'Annual Wages','MP_x', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 
                                                   'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]
EPL_2018_sum_complete_df.head()
```

```python
# Rename columns
EPL_2018_sum_complete_df = EPL_2018_sum_complete_df.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
EPL_2018_sum_complete_df.head()
```

```python
# Sort by league position
EPL_2018_sum_complete_df.sort_values('Pts', ascending=False)
```
