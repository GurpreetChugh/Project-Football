```python
import pandas as pd
import numpy as np
```

```python
# File to Load 
filepath = "Data/spi_matches_latest.csv"
# Read in csv file
df = pd.read_csv(filepath)
```

```python
#Get column names
df.columns
```

```python
#Create new DataFrame with desired columns
df2 = df[['season', 'date', 'league_id', 'league', 'team1', 'team2', 'score1', 'score2', 'xg1','xg2']]

#Display DataFrame
df2.sample(20)
```

```python
# Create DataFrame which shows only EPL matches
EPL_df = df2.loc[df2['league'] == 'Barclays Premier League']
EPL_df
```

```python
# Create DataFrame which shows only LaLiga matches
LaLiga_df = df2.loc[df2['league'] == 'Spanish Primera Division']
LaLiga_df
```

```python
#Rename Columns
EPL_cleaned_df = EPL_df.rename(columns={ 'team1': 'Home Team', 'team2': 'Away Team', 'score1': 'Home Score', 
                            'score2': 'Away Score', 'xg1': 'xGoals Home', 'xg2': 'xGoals Away'})
EPL_cleaned_df.head()
```

```python
#Drop all rows without xg data
EPL_cleaned_df.dropna(subset = ['xGoals Home'], inplace = True)
EPL_cleaned_df.dropna(subset = ['xGoals Away'], inplace = True)
EPL_cleaned_df
```

```python
#Rename Columns
LaLiga_cleaned_df = LaLiga_df.rename(columns={ 'team1': 'Home Team', 'team2': 'Away Team', 'score1': 'Home Score', 
                            'score2': 'Away Score', 'xg1': 'xGoals Home', 'xg2': 'xGoals Away'})
LaLiga_cleaned_df.head()
```

```python
#Drop all rows without xG data
LaLiga_cleaned_df.dropna(subset = ['xGoals Home'], inplace = True)
LaLiga_cleaned_df.dropna(subset = ['xGoals Away'], inplace = True)
LaLiga_cleaned_df
```
