```python
import pandas as pd
# from highlight_text import ax_text, fig_text
# from matplotlib.colors import LinearSegmentedColormap
import matplotlib.pyplot as plt
```

```python
#Get summary tables
laliga_18_sum = pd.read_html('https://fbref.com/en/comps/12/2017-2018/2017-2018-La-Liga-Stats')
laliga_19_sum = pd.read_html('https://fbref.com/en/comps/12/2018-2019/2018-2019-La-Liga-Stats')
laliga_20_sum = pd.read_html('https://fbref.com/en/comps/12/2019-2020/2019-2020-La-Liga-Stats')
laliga_21_sum = pd.read_html('https://fbref.com/en/comps/12/2020-2021/2020-2021-La-Liga-Stats')
laliga_22_sum = pd.read_html('https://fbref.com/en/comps/12/2021-2022/2021-2022-La-Liga-Stats')

LaLiga_2018_sum_df = pd.DataFrame(laliga_18_sum[0])
LaLiga_2019_sum_df = pd.DataFrame(laliga_19_sum[0])
LaLiga_2020_sum_df = pd.DataFrame(laliga_20_sum[0])
LaLiga_2021_sum_df = pd.DataFrame(laliga_21_sum[0])
LaLiga_2022_sum_df = pd.DataFrame(laliga_22_sum[0])
```

```python
# Get 2nd summary table (index[2])for each season
laliga_18_sum = pd.read_html('https://fbref.com/en/comps/12/2017-2018/2017-2018-La-Liga-Stats')
laliga_19_sum = pd.read_html('https://fbref.com/en/comps/12/2018-2019/2018-2019-La-Liga-Stats')
laliga_20_sum = pd.read_html('https://fbref.com/en/comps/12/2019-2020/2019-2020-La-Liga-Stats')
laliga_21_sum = pd.read_html('https://fbref.com/en/comps/12/2020-2021/2020-2021-La-Liga-Stats')
laliga_22_sum = pd.read_html('https://fbref.com/en/comps/12/2021-2022/2021-2022-La-Liga-Stats')

LaLiga_2018_sum2_df = pd.DataFrame(laliga_18_sum[2])
LaLiga_2019_sum2_df = pd.DataFrame(laliga_19_sum[2])
LaLiga_2020_sum2_df = pd.DataFrame(laliga_20_sum[2])
LaLiga_2021_sum2_df = pd.DataFrame(laliga_21_sum[2])
LaLiga_2022_sum2_df = pd.DataFrame(laliga_22_sum[2])
```

```python
# Remove upper level to allow for merging with other tables
LaLiga_2018_sum2_df = LaLiga_2018_sum2_df.droplevel(0, axis = 1)
LaLiga_2019_sum2_df = LaLiga_2019_sum2_df.droplevel(0, axis = 1)
LaLiga_2020_sum2_df = LaLiga_2020_sum2_df.droplevel(0, axis = 1)
LaLiga_2021_sum2_df = LaLiga_2021_sum2_df.droplevel(0, axis = 1)
LaLiga_2022_sum2_df = LaLiga_2022_sum2_df.droplevel(0, axis = 1)
```

```python
# Get wages table for each of the last 5 seasons
laliga_18wages = pd.read_html('https://fbref.com/en/comps/12/2017-2018/wages/2017-2018-La-Liga-Wages')
laliga_19wages = pd.read_html('https://fbref.com/en/comps/12/2018-2019/wages/2018-2019-La-Liga-Wages')
laliga_20wages = pd.read_html('https://fbref.com/en/comps/12/2019-2020/wages/2019-2020-La-Liga-Wages')
laliga_21wages = pd.read_html('https://fbref.com/en/comps/12/2020-2021/wages/2020-2021-La-Liga-Wages')
laliga_22wages = pd.read_html('https://fbref.com/en/comps/12/2021-2022/wages/2021-2022-La-Liga-Wages')

laliga_18wages_df = pd.DataFrame(laliga_18wages[0])
laliga_19wages_df = pd.DataFrame(laliga_19wages[0])
laliga_20wages_df = pd.DataFrame(laliga_20wages[0])
laliga_21wages_df = pd.DataFrame(laliga_21wages[0])
laliga_22wages_df = pd.DataFrame(laliga_22wages[0])
```

```python
# Merge wages and season summary for each season
LaLiga18_merged_df = pd.merge(laliga_18wages_df,LaLiga_2018_sum_df, how="left", on=['Squad'])
LaLiga19_merged_df = pd.merge(laliga_19wages_df,LaLiga_2019_sum_df, how="left", on=['Squad'])
LaLiga20_merged_df = pd.merge(laliga_20wages_df,LaLiga_2020_sum_df, how="left", on=['Squad'])
LaLiga21_merged_df = pd.merge(laliga_21wages_df,LaLiga_2021_sum_df, how="left", on=['Squad'])
LaLiga22_merged_df = pd.merge(laliga_22wages_df,LaLiga_2022_sum_df, how="left", on=['Squad'])

```

```python
# Merge in summary table#2 for each season
LaLiga18_complete_df = pd.merge(LaLiga18_merged_df, LaLiga_2018_sum2_df, how="left", on=['Squad'])
LaLiga19_complete_df = pd.merge(LaLiga19_merged_df, LaLiga_2019_sum2_df, how="left", on=['Squad'])
LaLiga20_complete_df = pd.merge(LaLiga20_merged_df, LaLiga_2020_sum2_df, how="left", on=['Squad'])
LaLiga21_complete_df = pd.merge(LaLiga21_merged_df, LaLiga_2021_sum2_df, how="left", on=['Squad'])
LaLiga22_complete_df = pd.merge(LaLiga22_merged_df, LaLiga_2022_sum2_df, how="left", on=['Squad'])
```

```python
#Get columns
LaLiga18_complete_df.columns
```

```python
# Select desired columns for final summary table
LaLiga18_sum_complete_df = LaLiga18_complete_df[['Rk_y', 'Squad','# Pl_x', 'Age', 'Poss', 'Annual Wages','MP_x', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 
                                                   'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

LaLiga19_sum_complete_df = LaLiga19_complete_df[['Rk_y', 'Squad','# Pl_x', 'Age', 'Poss', 'Annual Wages','MP_x', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 
                                                   'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

LaLiga20_sum_complete_df = LaLiga20_complete_df[['Rk_y', 'Squad','# Pl_x', 'Age', 'Poss', 'Annual Wages','MP_x', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 
                                                   'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

LaLiga21_sum_complete_df = LaLiga21_complete_df[['Rk_y', 'Squad','# Pl_x', 'Age', 'Poss', 'Annual Wages','MP_x', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 
                                                   'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

LaLiga22_sum_complete_df = LaLiga22_complete_df[['Rk_y', 'Squad','# Pl_x', 'Age', 'Poss', 'Annual Wages','MP_x', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 
                                                   'Pts/MP', 'xG_x', 'xGA', 'xGD', 'xGD/90', 'Attendance']]

```

```python
# Rename columns
LaLiga18_sum_complete_df = LaLiga18_complete_df.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
LaLiga19_sum_complete_df = LaLiga19_complete_df.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
LaLiga20_sum_complete_df = LaLiga20_complete_df.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
LaLiga21_sum_complete_df = LaLiga21_complete_df.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
LaLiga22_sum_complete_df = LaLiga22_complete_df.rename(columns={ 'Rk_y': 'League Rank', 'Age': 'Avg Age', 
                                                                    'Poss': 'Avg Poss', 
                                                                    '# Pl_x': '# of Players', 'MP_x': 'MP', 
                                                                    'xG_x': 'xG'})
```

```python
# Convert 'Annual Wages' column into single column with only dollar value
LaLiga18_sum_complete_df[['pounds', 'dollars']] = LaLiga18_sum_complete_df['Annual Wages'].str.split('$', expand=True)
LaLiga19_sum_complete_df[['pounds', 'dollars']] = LaLiga19_sum_complete_df['Annual Wages'].str.split('$', expand=True)
LaLiga20_sum_complete_df[['pounds', 'dollars']] = LaLiga20_sum_complete_df['Annual Wages'].str.split('$', expand=True)
LaLiga21_sum_complete_df[['pounds', 'dollars']] = LaLiga21_sum_complete_df['Annual Wages'].str.split('$', expand=True)
LaLiga22_sum_complete_df[['pounds', 'dollars']] = LaLiga22_sum_complete_df['Annual Wages'].str.split('$', expand=True)
```

```python
# Convert 'Annual Wages' column into single column with only dollar value
LaLiga18_sum_complete_df[['Annual Wages ($)', 'discard']] = LaLiga18_sum_complete_df['dollars'].str.split(')', expand=True)
LaLiga19_sum_complete_df[['Annual Wages ($)', 'discard']] = LaLiga19_sum_complete_df['dollars'].str.split(')', expand=True)
LaLiga20_sum_complete_df[['Annual Wages ($)', 'discard']] = LaLiga20_sum_complete_df['dollars'].str.split(')', expand=True)
LaLiga21_sum_complete_df[['Annual Wages ($)', 'discard']] = LaLiga21_sum_complete_df['dollars'].str.split(')', expand=True)
LaLiga22_sum_complete_df[['Annual Wages ($)', 'discard']] = LaLiga22_sum_complete_df['dollars'].str.split(')', expand=True)

```

```python
# Drop unwanted columns 
LaLiga18_sum_clean_df = LaLiga18_sum_complete_df.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
LaLiga19_sum_clean_df = LaLiga19_sum_complete_df.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
LaLiga20_sum_clean_df = LaLiga20_sum_complete_df.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
LaLiga21_sum_clean_df = LaLiga21_sum_complete_df.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)
LaLiga22_sum_clean_df = LaLiga22_sum_complete_df.drop(['pounds', 'dollars', 'Annual Wages', 'discard'], axis=1)

```

```python
# Sort by league position
LaLiga18_sum_clean_sorted_df = LaLiga18_sum_clean_df.sort_values('Pts', ascending=False)
LaLiga19_sum_clean_sorted_df = LaLiga19_sum_clean_df.sort_values('Pts', ascending=False)
LaLiga20_sum_clean_sorted_df = LaLiga20_sum_clean_df.sort_values('Pts', ascending=False)
LaLiga21_sum_clean_sorted_df = LaLiga21_sum_clean_df.sort_values('Pts', ascending=False)
LaLiga22_sum_clean_sorted_df = LaLiga22_sum_clean_df.sort_values('Pts', ascending=False)
```

```python
LaLiga18_sum_clean_sorted_df['Season'] = '2017-2018'
LaLiga19_sum_clean_sorted_df['Season'] = '2018-2019'
LaLiga20_sum_clean_sorted_df['Season'] = '2019-2020'
LaLiga21_sum_clean_sorted_df['Season'] = '2020-2021'
LaLiga22_sum_clean_sorted_df['Season'] = '2021-2022'
```

```python
# Concatenate all seasons into one dataframe
LaLiga_allseasons_sum_df = pd.concat([LaLiga18_sum_clean_sorted_df, LaLiga19_sum_clean_sorted_df, LaLiga20_sum_clean_sorted_df,
LaLiga21_sum_clean_sorted_df, LaLiga22_sum_clean_sorted_df], axis=0)

LaLiga_allseasons_sum_df.info()
```
